```python
import pandas as pd, csv

path = "car_sales_cleaned1.csv"
encodings = ["utf-8-sig", "utf-8", "cp1251", "latin1"]
seps = [",", ";", "\t", "|"]

best = None
for enc in encodings:
    for sep in seps:
        try:
            df_try = pd.read_csv(path, sep=sep, encoding=enc, nrows=500)
            # –µ—Å–ª–∏ —Ä–µ–∞–ª—å–Ω–æ –Ω–µ—Å–∫–æ–ª—å–∫–æ –∫–æ–ª–æ–Ω–æ–∫ ‚Äî —Å—á–∏—Ç–∞–µ–º, —á—Ç–æ –Ω–∞—à–ª–∏
            if df_try.shape[1] > 1:
                best = (enc, sep)
                break
        except Exception:
            pass
    if best:
        break

if best:
    enc, sep = best
    # –ø–æ–≤—Ç–æ—Ä–Ω–æ —á–∏—Ç–∞–µ–º –≤–µ—Å—å —Ñ–∞–π–ª –∏ —É—á–∏—Ç—ã–≤–∞–µ–º –≤–æ–∑–º–æ–∂–Ω—É—é –∑–∞–ø—è—Ç—É—é –∫–∞–∫ –¥–µ—Å—è—Ç–∏—á–Ω—ã–π —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å
    try:
        df = pd.read_csv(path, sep=sep, encoding=enc, decimal=",", thousands=".")
    except Exception:
        df = pd.read_csv(path, sep=sep, encoding=enc)
    print(f"‚úÖ OK: encoding={enc}, sep={repr(sep)}, shape={df.shape}")
    display(df.head())
    display(df.tail())
else:
    print("‚ùå –ù–µ —É–¥–∞–ª–æ—Å—å –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏ —Ä–∞—Å–ø–∞—Ä—Å–∏—Ç—å. –ü–æ–ø—Ä–æ–±—É–π –ø–µ—Ä–µ—Å–æ—Ö—Ä–∞–Ω–∏—Ç—å —Ñ–∞–π–ª –∫–∞–∫ 'CSV UTF-8 (Comma delimited)'.")
```

```python
import pandas as pd

# –ó–∞–≥—Ä—É–∂–∞–µ–º –∫–∞–∫ —Ç–µ–∫—Å—Ç –∏ —á–∏—Å—Ç–∏–º –∫–∞–≤—ã—á–∫–∏
df_raw = pd.read_csv("car_sales_cleaned1.csv", header=None, encoding="utf-8")

# –†–∞–∑–±–∏–≤–∞–µ–º —Å—Ç—Ä–æ–∫—É –Ω–∞ –∫–æ–ª–æ–Ω–∫–∏
df = df_raw[0].str.replace('"', '').str.split(",", expand=True)

# –ó–∞–¥–∞—ë–º –∏–º–µ–Ω–∞ –∫–æ–ª–æ–Ω–æ–∫
df.columns = ['Gender', 'Annual Income', 'Company', 'Model', 'Transmission', 'Color', 'Price ($)']

# –°–æ—Ö—Ä–∞–Ω—è–µ–º —á–∏—Å—Ç—É—é –≤–µ—Ä—Å–∏—é
df.to_csv("clean_cars.csv", index=False, encoding="utf-8")
print("‚úÖ –§–∞–π–ª —Å–æ—Ö—Ä–∞–Ω—ë–Ω –∫–∞–∫ clean_cars.csv")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import joblib

# –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏ —Ü–µ–ª–∏
X = df[["Gender", "Color"]].copy()
y = df["Model"].copy()

le = LabelEncoder()
for col in X.columns:
    X[col] = le.fit_transform(X[col])
y = le.fit_transform(y)

# –î–µ–ª–∏–º –Ω–∞ train/test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# –û–±—É—á–∞–µ–º
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# –ü—Ä–æ–≤–µ—Ä—è–µ–º —Ç–æ—á–Ω–æ—Å—Ç—å
y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f"üéØ Accuracy: {acc:.3f}")

# –°–æ—Ö—Ä–∞–Ω—è–µ–º –º–æ–¥–µ–ª—å
joblib.dump(model, "car_model_predictor.pkl")
print("üíæ –ú–æ–¥–µ–ª—å —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∞ –∫–∞–∫ car_model_predictor.pkl")
```

```python
# –¢–æ–ª—å–∫–æ Hyundai
df_hyundai = df[df["Company"] == "Hyundai"]
print("Hyundai sample:")
print(df_hyundai.head())

# –¢–æ–ª—å–∫–æ Toyota —Å —Ü–µ–Ω–æ–π > 40000
df_toyota = df[(df["Company"] == "Toyota") & (df["Price ($)"].astype(float) > 40000)]
print("\nToyota > 40000 sample:")
print(df_toyota.head())

# –¢–æ–ø-3 –º–æ–¥–µ–ª–∏
top3 = df["Model"].value_counts().head(3)
print("\nüî• Top-3 car models:")
print(top3)
```

```python
# –£–±–∏—Ä–∞–µ–º –∑–∞–≥–æ–ª–æ–≤–æ–∫, –ø—Ä–æ–±–µ–ª—ã, —Å–∏–º–≤–æ–ª—ã $, –∑–∞–ø—è—Ç—ã–µ
df["Price ($)"] = (
    df["Price ($)"]
    .str.replace("Price ($)", "", regex=False)
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.strip()
)

# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –≤ float, –Ω–µ—á–∏—Å–ª–æ–≤—ã–µ ‚Äî –≤ NaN
df["Price ($)"] = pd.to_numeric(df["Price ($)"], errors="coerce")
```
